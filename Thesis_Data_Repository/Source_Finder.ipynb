```python
search_queries = [
   "AI", "professional cycling", "Cycing training tactics", "Cycling performance analytics",
"Machine learning", "cycling performance", "Cycling data analysis", "athlete monitoring",
"Algorithms", "cycling training", "Cycling optimisation", "strategy in cycling", "Data analytics",
"World Tour cycling", "team strategy", "rider development", "Predictive analytics within cycling",
"cycling", "race outcomes", "injury prevention", "cycling aerodynamics",
"bike design", "power output", "cycling tactics", "cycling race simulation",
"AI applications within cycling", "cycling nutrition", "Cycling diet planning", "Cycling recovery", "Computer vision",
"cycling data", "AI biomechanics", "AI technique analysis", "AI injury prevention",
"predictive risk assessment", "AI used for rehabilitation", "Artificial intelligence", "cycling power meters",
"data interpretation within cycling", "training zones", "pedal stroke analysis", "cycling fficiency",
"Sports science technology", "innovation within cycing", "advanced tools in cycling", "cycling team",
"strategic planning", "Is AI a competitive advantage", "Big data", "cycling sport",
"cycling trends", "Digital innovation", "cycling professional",
"technology adoption", "future of cycling", "AI driven developments within cycling", "AI powered cycling", "AI tools within cycling"


]

!pip install ddgs

from ddgs import DDGS

collected_urls = set()
ddgs = DDGS()
max_urls_to_collect = 1000

print(f"Starting URL collection with {len(search_queries)} queries...")

for query in search_queries:
    print(f"Searching for: '{query}'")
    try:
        # Corrected method call: pass 'query' directly as the positional argument
        results = ddgs.text(query, max_results=100)
        for result in results:
            if 'href' in result and result['href']:
                collected_urls.add(result['href'])
            if len(collected_urls) >= max_urls_to_collect:
                print(f"Reached {max_urls_to_collect} unique URLs. Stopping search.")
                break
    except Exception as e:
        print(f"An error occurred while searching for '{query}': {e}")

    if len(collected_urls) >= max_urls_to_collect:
        break

print(f"\nTotal number of unique URLs collected: {len(collected_urls)}")
print("First 5 collected URLs:")
for url in list(collected_urls)[:5]:
    print(f"- {url}")
```

```python
print(f"Total collected URLs: {len(collected_urls)}")
for i, url in enumerate(list(collected_urls)):
    print(f"- {url}")
    if i >= 19: # Displaying up to 20 URLs
        print(f"... and {len(collected_urls) - 20} more URLs.")
        break
```

```python
import pandas as pd

# Convert the set of URLs to a list
urls_list = list(collected_urls)

# Create a pandas DataFrame from the list of URLs
df_urls = pd.DataFrame(urls_list, columns=['URL'])

# Save the DataFrame to a CSV file
df_urls.to_csv('collected_urls.csv', index=False)

print(f"Successfully saved {len(urls_list)} URLs to collected_urls.csv")
```
